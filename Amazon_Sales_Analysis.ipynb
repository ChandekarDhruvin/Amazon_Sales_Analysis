#Size

```python
size_count =sns.countplot(data = df,x = "Size",hue = "Size")

for cnt in size_count.containers:
  size_count.bar_label(cnt)
```

```python
#How many quantity they are buying
buying_quantity = df.groupby(['Size'],as_index = False)['Quantity'].sum().sort_values(by = "Quantity",ascending = False)
sns.barplot(data = buying_quantity,x = 'Size',y = "Quantity",hue = "Size")
```

#**Most of the buyers are bying M Size Clothes**

#Courier

```python
# df["Courier Status"]
# plt.figure(figsize= (10,5))
courier_status_count = sns.countplot(data = df,x = "Courier Status",hue = "Status")
# for cnt in courier_status_count.containers:
#   courier_status_count.bar_label(cnt)
```

```python
#Quantity shipped
quantity_shipped = df.groupby(["Courier Status"],as_index = False)['Quantity'].sum().sort_values(by = 'Quantity',ascending = False)
x = sns.barplot(data= quantity_shipped,x = "Courier Status",y = "Quantity",hue = "Courier Status")
for cnt in x.containers:
  x.bar_label(cnt)
```

#**Most of the orders are Shipped**

#Category

```python
df.info()
plt.figure(figsize = (10,5))
df['Category'] = df['Category'].astype('str')
cloths_category = df['Category']
plt.hist(cloths_category,bins = 30,color = "pink",edgecolor = 'Red')
plt.xticks(rotation = 90)
plt.title("Most of The Bought Cloths")
plt.show()
```

```python
plt.figure(figsize = (10,5))
category_count = sns.countplot(data = df,x = "Category",hue = "Category")
for x in category_count.containers:
  category_count.bar_label(x)
```

#**Most of the buyers are buying T-shirts**

#B2B

```python
B2B_dt = df['B2B'].value_counts()
B2B_dt
```

```python
plt.pie(B2B_dt,labels = B2B_dt,autopct = '%1.1f%%')
plt.show()
```

#**99.2% Buyers are Retailers and 0.8% Buyers are B2B**

#Top 10 states Distribution Wise

```python
top_10_states = df['ship-state'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.countplot(data = df[df['ship-state'].isin(top_10_states.index)], x= 'ship-state',hue = 'ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('State Distibution')
plt.xticks(rotation = 45)
plt.show()
```

#**Most of the buyers are from Maharashtra State**
